# Q10 

```python
nd=df.loc[:,['Country_Region','Confirmed','Deaths', 'Recovered', 'Active']]
nd=nd[nd.Deaths>150].reset_index(drop=True)
nd
```

```python
plt.plot(nd.Country_Region,nd.Deaths,color='r', linestyle='--',marker='.' ,linewidth=3,label='Deaths')
plt.plot(nd.Country_Region,nd.Recovered, color='b',marker='o',label = 'Recovered')
plt.plot(nd.Country_Region,nd.Active, color='y',marker='o',label = 'Active')
plt.plot(nd.Country_Region,nd.Confirmed, color='k',marker='o',label = 'Confirmed')
plt.title('Country wise comparison')
plt.xlabel('Country Region')
plt.ylabel('avg_ratio')
plt.legend()
plt.xticks(rotation='vertical')
plt.grid(True)
plt.tight_layout(30,30)
plt.show()
```

# Q11

```python
nd = df[df['Country_Region']=='US'].drop(['Country_Region'], axis=1)
nd = nd[nd.sum(axis = 1) > 0]
nd = nd.groupby(['Province_State'])['Deaths'].sum().reset_index()
us_data_death = nd[nd['Deaths'] > 0]
state_fig = px.bar(us_data_death, x='Province_State', y='Deaths', title='State wise deaths reported of COVID-19 in USA', text='Deaths')
state_fig.show()
```

# Q12 

```python
df['Active'] = df['Confirmed'] - df['Deaths'] - df['Recovered']
nd = df[df['Country_Region']=='US'].drop(['Country_Region',], axis=1)
nd = nd[nd.sum(axis = 1) > 0]
 
nd = nd.groupby(['Province_State'])['Active'].sum().reset_index()
us_data_death = nd[nd['Active'] > 0]
state_fig = px.bar(us_data_death, x='Province_State', y='Active', title='State wise recovery cases of COVID-19 in USA', text='Active')
state_fig.show()
```

# Q13

```python
df['Active'] = df['Confirmed'] - df['Deaths'] - df['Recovered']
pk_data = df[df['Country_Region']=='Pakistan'].drop(['Country_Region',], axis=1)
pk_data = pk_data[pk_data.sum(axis = 1) > 0]
pk_data = pk_data.groupby(['Province_State'])['Confirmed', 'Deaths', 'Recovered', 'Active'].sum().reset_index()
pk_data = pd.melt(pk_data, id_vars='Province_State', value_vars=['Confirmed', 'Deaths', 'Recovered', 'Active'], value_name='Count', var_name='Case')
fig = px.bar(pk_data, x='Province_State', y='Count', text='Count', barmode='group', color='Case', title='Pakistan State wise combine number of confirmed, deaths, recovered, active COVID-19 cases')
fig.show()
```
